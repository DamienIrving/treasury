```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
infiles = {
    'ACCESS-CM2': '/g/data/xv83/dbi599/treasury/WSDI/ACCESS-CM2/ssp370/wsdi_yr_ACCESS-CM2_ssp370_ensemble_aus-states-cities_1850-2100.csv',
    'MPI-ESM1-2-LR' : '/g/data/xv83/dbi599/treasury/WSDI/MPI-ESM1-2-LR/ssp370/wsdi_yr_MPI-ESM1-2-LR_ssp370_ensemble_aus-states-cities_1850-2100.csv',
}
```

```python
df_dict = {}
df_dict['ACCESS-CM2'] = pd.read_csv(infiles['ACCESS-CM2'])
df_dict['MPI-ESM1-2-LR'] = pd.read_csv(infiles['MPI-ESM1-2-LR'])
```

```python
df_dict['ACCESS-CM2']
```

```python
def plot_location(model, location, start_year=1950):
    """Scatterplot for a given location"""

    df = df_dict[model]
    fig, ax = plt.subplots(figsize=[15,7])
    sns.scatterplot(
        ax=ax,
        data=df[df['year'] >= start_year],
        x='year',
        y=location,
        hue='run',
    )
    ax.set_ylabel('WSDI')
    ax.set_title(f'{model} - {location}')
    ax.grid()
    plt.show()
```

```python
plot_location('ACCESS-CM2', 'AUS')
```

```python
plot_location('MPI-ESM1-2-LR', 'AUS')
```

```python
plot_location('ACCESS-CM2', 'Melbourne')
```

```python
plot_location('MPI-ESM1-2-LR', 'Melbourne')
```
