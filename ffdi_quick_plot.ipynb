```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
infiles = {}
infiles['FFDIx'] = {
    'ACCESS-ESM1-5': '/g/data/xv83/dbi599/treasury/FFDI/ACCESS-ESM1-5/ssp370/FFDIx_yr_ACCESS-ESM1-5_ssp370_r1i1p1f1_aus-states_1850-2100.csv',
}
infiles['FFDIgt99p'] = {
    'ACCESS-ESM1-5': '/g/data/xv83/dbi599/treasury/FFDI/ACCESS-ESM1-5/ssp370/FFDIgt99p_yr_ACCESS-ESM1-5_ssp370_r1i1p1f1_aus-states_1850-2100.csv',
}
```

```python
df_dict = {}
df_dict['FFDIx'] = { 
    'ACCESS-ESM1-5': pd.read_csv(infiles['FFDIx']['ACCESS-ESM1-5']),
}
df_dict['FFDIgt99p'] = { 
    'ACCESS-ESM1-5': pd.read_csv(infiles['FFDIgt99p']['ACCESS-ESM1-5']),
}
```

```python
df_dict['FFDIgt99p']['ACCESS-ESM1-5']
```

```python
def plot_location(metric, model, location, start_year=1950):
    """Scatterplot for a given location"""

    df = df_dict[metric][model]
    fig, ax = plt.subplots(figsize=[15,7])
    sns.scatterplot(
        ax=ax,
        data=df[df['year'] >= start_year],
        x='year',
        y=location,
        hue='run',
    )
    ax.set_ylabel(metric)
    ax.set_title(f'{model} - {location}')
    ax.grid()
    plt.show()
```

```python
plot_location('FFDIx', 'ACCESS-ESM1-5', 'AUS')
```

```python
plot_location('FFDIx', 'ACCESS-ESM1-5', 'SA')
```

```python
plot_location('FFDIgt99p', 'ACCESS-ESM1-5', 'AUS')
```

```python
plot_location('FFDIgt99p', 'ACCESS-ESM1-5', 'SA')
```
